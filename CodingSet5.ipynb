### Problem 16



#### High-Level Overview

I started by creating the is_prime function. By iterating through every number between 2 and the number, we can find if it is divisible by any of them. If it isn't, it's a prime number. Then, outside of the function, I decided to loop through to whenever the user chooses to quit while it asks the user for input.


```python
#prompt the user for input
while True:
    user = input("Enter a number (type 'x' to quit): ")
    if user.lower() == 'x':
        break
    print(is_prime(int(user)))
    
#define a function called is_prime that finds prime numbers
def is_prime(num):
    #find if the number isn't a prime number through modulo
    for i in range(2,num):
        if (num % i) == 0:
            return False
    #return true since the number isn't divisible by any numbers other than it's self.
    return True
```

### Problem 18



#### High-Level Overview

Originally, I defined the function after the user input, this caused an error so I rearranged it. Other than that, I just had to make sure that my equation was right by adding parentheses. It is also important to use float inputs for the present value and the interest to get accurate results.


```python
#define a function that returns the future value of the account after the amount of months specified.
def future_amount(present, interest, months):
    futureA = present * (1 + interest)**months
    print(f"Future Amount: {futureA: .2f}")

#ask user for input
present = float(input("Enter your account's present value: "))
interest = float(input("Enter your monthly interest rate: "))
months = int(input("Enter the amount of months that the money will be left in the account: "))

#call function
future_amount(present, interest, months)
```
