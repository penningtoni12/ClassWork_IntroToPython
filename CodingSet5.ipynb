### Problem 16



#### High-Level Overview

I started by creating the is_prime function. By iterating through every number between 2 and the number, we can find if it is divisible by any of them. If it isn't, it's a prime number. Then, outside of the function, I decided to loop through to whenever the user chooses to quit while it asks the user for input.


```python
#prompt the user for input
while True:
    user = input("Enter a number (type 'x' to quit): ")
    if user.lower() == 'x':
        break
    print(is_prime(int(user)))
    
#define a function called is_prime that finds prime numbers
def is_prime(num):
    #find if the number isn't a prime number through modulo
    for i in range(2,num):
        if (num % i) == 0:
            return False
    #return true since the number isn't divisible by any numbers other than it's self.
    return True
```
