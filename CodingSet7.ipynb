### Problem 1



#### High-Level Overview

This code should start by creating a list, and then iterating through the days of the week to get the expense values. Make sure that the loop handles value errors. Then, add the values the user inputs to the list. Use the sum built-in function to add all of the elements together.


```python
# create a main function
def main():
    # use try/except clauses to prevent anything except the correct values from being added to the list.
    # First, start by using a loop to ask the user for their expenses for the week
    #Also, create a list
    sales = []
    #iterate 7 times for each day of the week
    for i in range(7):
        while True:
            try:
                #ask the user for their daily expenses
                expense = float(input(f'What was your total sales for day {i+1}?'))
                sales.append(expense)
                break
            except ValueError:
                print("Please input a valid number.")
    #add the values in the list
    
    total_sales = sum(sales)
    print(f'The total sales for the week was {total_sales: .2f}$')
    
if __name__ == "__main__":
    main()
```

### Problem 14



#### High-Level Overview

The hardest part of this code is getting the loop correct. Once, you are able to get the special characters off of the end of the values, you can make a pie chart by importing the matplotlib at the beginning of the code. I added try and excpet clauses just in case there wasn't a file there.


```python
#import matplot
import matplotlib.pyplot as plt
#create the main function
def main():
    try:
        #open expense file
        with open("expenses.txt") as expense_file:
            # make a list with the expenses as it's elements
            expenses = expense_file.readlines()
            #strip the special character from the values in expenses
            for i in range(len(expenses)):
                expenses[i] = expenses[i].strip()
        #label the parts of the pie chart
        slice_labels = ["Rent", "Gas", "Food", "Clothing", "Car Payment", "Misc"]
        plt.pie(expenses, labels=slice_labels)
        plt.title("Last Month's Expenses")
        #display
        plt.show()
    except FileNotFoundError:
        print("This file does not exist")

if __name__ == "__main__":
    main()
```
