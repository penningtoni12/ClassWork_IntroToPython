### Problem 3



#### High-Level Overview

This code wass pretty fun. I thought it would be easiest to separate the month day and year into a list, this would make them easier to manipulate. I made a list of month's so that it would be easier to use in the print statement. The date_changer function splits the date inputted, converts the month into an int and then print's the month correlated with the number, then joins the rest of the split day and year.


```python
# define main function 
def main():
    day_input = input("Enter a date in the format mm/dd/yyyy: ")
    date_changer(day_input)

#make a function that replaces '/' with ',' and numbered months with their names.
def date_changer(date):
    months = ["January", "Febuary", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
    separate_dates = date.split('/')
    try:
        num = int(separate_dates[0])
        print(f"{months[num-1]} {separate_dates[1]}, {separate_dates[2]}.")
    except ValueError:
        print("There was an error calculating the month")


if __name__ == "__main__":
    main()
```

```python
#making the alphabet, numbers, and punctuation global
alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
numbers = "123456789"
punctuation = [" ", ",", ".", "?"]
#make a main function
def main():
    user_input = input("Enter text to be translated to morse: ")
    morse(user_input)

#make a function that converts characters to morse code
def morse(text):
    morse_text = [".-", "-...", "-.-.", "-..", ".", "..-.", "--.", "....", "..", ".---", "-.-", ".-..", "--", "-.", "---", ".--.", "--.-", ".-.", "...", "-", "..-", "...-", ".--", "-..-", "-.-", "--.."]
    morse_num = ["-----", ".---", "..---", "...--", "....-", ".....", "-....", "--...", "---..", "----."]
    morse_punc = [" ", "--..--", ".-.-.-", "..--.."]
    morse_encoded = []
    for char in text:
        if char.upper() in alphabet:
            spot = alphabet.index(char.upper())
            morse_encoded.append(morse_text[spot])

        elif char in numbers:
            spot_num = numbers.index(char)
            morse_encoded.append(morse_num[spot_num])

        elif char in punctuation:
            spot_punc = punctuation.index(char)
            morse_encoded.append(morse_punc[spot_punc])

    for letter in morse_encoded:
        print(f"{letter}", end="|")



if __name__ == "__main__":
    main()
```
