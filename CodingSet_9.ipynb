```python
# create a dictionary with values corresponding to the alphebet to be coded
# make this global so that it can be used in all functions*
codes = {'A':'%', 'a':'9', 'B':'@', 'b':'#', 'C':'!', 'c':'s', 'D':'$', 'd':'-', 'E':'0', 'e':')', 'F':'^', 'f':'*', 'G':'7', 'g':'5', 'H':'(', 'h':'_', 'I':'+', 'i':'=', 'J':'`', 'j':'~', 'K':',', 'k':'<', 'L':'>', 'l':'/', 'M':'?', 'm':';', 'N':':', 'n':'"', 'O':'[', 'o':']', 'P':'{', 'p':'}', 'Q':'3', 'q':'2', 'R':'4', 'r':'8', 'S':'6', 's':'1', 'T':'|', 't':'a', 'U':'e', 'u':'n', 'V':'i', 'v':'L', 'W':'m', 'w':'v', 'X':'q', 'x':'t', 'Y':'z', 'y':'h', 'Z':'d', 'z':'p', ' ':' ', '.':'.'} 


#make a main function that calls on the functions
def main():
    # open a file and call the code_encrypter function to translate it into code
    # make a try/except clause in case there are any problems opening or reading the file
    try:
        with open('Chapter9code.txt', 'r') as infile:
            # read the lines of the text file and strip any special character that is at the end
            text = infile.read().rstrip('\n')
            #call the encoder function
            code_encrypter(text)
        print()
        # open the file and call the code_decrypter function
        with open('Chapter_9_encodedText.txt', 'r') as infile:
            lang = infile.read().rstrip('\n')
            # call the function
            code_decrypter(lang)

    except FileNotFoundError:
        print("The file was not found!")

    except IOError:
        print("There was an error reading the code!")


# make a function that encrypts the message 
# use a conditon to be inputted when calling the function
def code_encrypter(string):
    # make an empty list that we can append values to
    coded = []
    # iterate through the string and append the values corresponding to the key that is found in the string
    for char in string:
        #append the values to the list
        coded.append(codes.get(char, '-1'))
        # I made the code print -1 in case there is a character not found in the dictionary

    #print the values of the list
    for text in coded:
        print(f'{text}', end="") # the end condition is there so that it will print on one line
        


# make a function that decodes any encoded message
def code_decrypter(message):
    # make an empty list so that values may be appended to it
    decrypted = []
    # this function needs to print the keys, not the values of the dictionary
    # the easies way to do this is use the get function.
    # If we reverse the keys and values, we can use the get function to obtain the values (this is what is the keys right now in the codes dictionary)
    reversed = {value:keys for keys, value in codes.items()}

    # iterate thrugh the message used in the function and append the values of the reversed dictionary to the decrypted list.
    for char in message:
        decrypted.append(reversed.get(char,'x')) # I added the 'x' for a failsafe in case the value isn't found

    #print the values appended
    for letter in decrypted:
        print(f'{letter}', end="")




if __name__ == "__main__":
    main()

#testing the code
```
